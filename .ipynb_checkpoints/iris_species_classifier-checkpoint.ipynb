### splitinf data into traning set and testing set:
### (X=>data, Y=>label)

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(iris['data'],iris['target'],random_state=0)
```

```python
len(X_train),len(X_test),len(Y_train),len(Y_test)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig,ax = plt.subplots(3,3,figsize=(15,15))
plt.suptitle("IRIS FEATURES PAIRPLOT")

for i in range(3):
    for j in range(3):
        ax[i,j].scatter(X_train[:,j],X_train[:,i+1],c=Y_train,s=60)
        ax[i,j].set_xticks(())
        ax[i,j].set_yticks(())
        if i==2:
            ax[i,j].set_xlabel(iris['feature_names'][j])
        if j==0:
            ax[i,j].set_ylabel(iris['feature_names'][i+1])
        if j>i:
            ax[i,j].set_visible(False)
```

### Above  plots shows three classes are well separated. so KNN will likely be able to learn to separate them.

### For k = 1:

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,Y_train)
```

### Now to make prediction, we can use predict method.we can directly use score() to find accuracy

```python
knn.score(X_test,Y_test)
```

### So accuracy is 97%

### Findig accuracy without score()

```python
import numpy as np

Y_predict = knn.predict(X_test)
np.mean(Y_predict==Y_test)
```

### predict for input: sepal length=4,sepal width=4,petal length=2,petal width=0.2

```python
X_predict = np.array([[4,4,2,.2]])
prediction = knn.predict(X_predict)
#X_predict.shape
prediction
```
